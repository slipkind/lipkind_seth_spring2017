```python
def preprocess(filename):
    file_lower = []
    file = webtext.words(filename)
    for string in file:
        file_lower.append(string.lower())
    return file_lower
```

```python
#s.translate(None, string.punctuation) for s in test_case]
def removeStopandPunc(nestedList):
    counts = dict()
    
    # for loop to search through persuasion
    for lists in nestedList:
        for word in lists:
            word2 = ''.join((x for x in word if word not in string.punctuation))
            if word2 not in stopwords.words('english'):
                for num in num_list:
                    if not (num in word2) and not (word2 == ''):
                        counts[word2] = counts.get(word2, 0) + 1 # get word and add value to count
    return counts
```
